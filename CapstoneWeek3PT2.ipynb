```python
import pandas as pd
```

```python
page = 'https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'
myTable = pd.read_html(page, attrs={"class":"wikitable"})
```

```python
df = myTable[0]
```

```python
df.head()
```

```python
good_df = df[df.Borough != 'Not assigned']
good_df = good_df.reset_index(drop=True)
```

```python
good_df.head(10)
```

```python
good_df['Neighbourhood'] = good_df.apply(lambda x: x['Borough'] if x['Neighbourhood']=='Not assigned' else x['Neighbourhood'], axis=1)
```

```python
good_df.head(50)
```

```python
grouped_df = good_df.groupby(['Postcode','Borough'])['Neighbourhood'].apply(lambda x: "%s" % ', '.join(x))
```

```python
grouped_df = grouped_df.to_frame()
```

```python
grouped_df.rename(columns={ df.columns[2]: "Neighbourhood" }, inplace = True)
```
